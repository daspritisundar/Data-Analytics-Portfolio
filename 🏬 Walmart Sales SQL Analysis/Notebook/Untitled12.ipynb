6.  Add a new secret with name `KAGGLE_USERNAME` and value of your kaggle username.
7.  Add a new secret with name `KAGGLE_KEY` and value of your kaggle key.

```python
#no. of duplicates
df.duplicated().sum()
```

```python
#delete duplicate entries
df = df.drop_duplicates()
#verify
df.duplicated().sum()
```

```python
#no. of null values
df.isnull().sum()
```

```python
#dropping all raws with missing values
df = df.dropna()
#verify
df.isnull().sum()
```

```python
df.shape
```

```python
#converting unit_proce column into float value
df['unit_price']=df['unit_price'].str.replace('$', '').astype(float)
```

```python
#creating a new column in the dataset
df['total_price'] = df['quantity'] * df['unit_price']

df.head()
```
